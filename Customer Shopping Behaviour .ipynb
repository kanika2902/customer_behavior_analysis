```python
from sqlalchemy import text

with engine.connect() as conn:
    rows = conn.execute(text(
        """
        SELECT s.name AS schema_name,
               t.name AS table_name,
               u.name AS owner_name,
               t.create_date
        FROM sys.tables t
        JOIN sys.schemas s ON t.schema_id = s.schema_id
        JOIN sys.database_principals u ON t.principal_id = u.principal_id
        ORDER BY t.create_date DESC;
        """
    )).fetchall()

for r in rows:
    print(r)
```

```python
from sqlalchemy import event, text
import pyodbc

# enable fast_executemany on pyodbc cursors (if not already set)
@event.listens_for(engine, "before_cursor_execute")
def _enable_fast_executemany(conn, cursor, statement, parameters, context, executemany):
    # set fast_executemany only for executemany operations
    if executemany:
        try:
            cursor.fast_executemany = True
        except Exception:
            pass

# write in reasonable chunks (adjust chunksize if you like)
df.to_sql(
    'customer',
    engine,
    schema='dbo',
    if_exists='replace',
    index=False,
    chunksize=1000,   # smaller batches avoid parameter explosion
    method=None       # IMPORTANT: don't use 'multi' here
)

# verify
with engine.connect() as conn:
    print(conn.execute(text("SELECT COUNT(*) FROM dbo.customer")).fetchone())
```
